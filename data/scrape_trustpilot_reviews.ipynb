```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import datetime as dt
import time
```

```python
# Set Trustpilot page numbers to scrape here
from_page = 1
to_page = 220

# Set the company
company = "if.fi"
```

```python
def get_reviews(company, from_page, to_page):
    reviews = []

    for i in range(from_page, to_page+1):
        time.sleep(3)
        URL = f"https://fi.trustpilot.com/review/{company}?languages=all&page={i}"
        print(URL)
        r = requests.get(URL)
        soup = BeautifulSoup(r.content, 'lxml')
        table = soup.findAll("section", {'class':'styles_reviewContentwrapper__zH_9M'})
    
        for row in range(len(table)):
            review = {}
            review['company'] = company
            review['rating'] = table[row].div['data-service-review-rating']
            review['time'] = table[row].div.time['datetime']
            review['title'] = table[row].h2.text
            review['review_text'] = table[row].p.text
            reviews.append(review)
        
    return reviews
```

```python
reviews = get_reviews(company, from_page, to_page)
reviews = pd.json_normalize(reviews)
reviews.to_csv(f'{company}_reviews.csv', encoding='utf-16', sep=';', errors='replace')
```
